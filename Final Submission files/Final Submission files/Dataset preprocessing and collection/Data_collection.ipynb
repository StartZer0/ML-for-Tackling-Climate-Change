#ERA5 - Required 2 variable collection from this dataset

We dowload the first set od data in chaos bench which has 2 of the required variables out of 10 we need

```python
!cd ChaosBench/data/ && ./process.sh era5
```

```python
!pip install xarray zarr netCDF4 fsspec
```

```python
import os

data_dir = "ChaosBench/data/era5"
all_files = sorted(os.listdir(data_dir))  # List all available files

# Filter only the files for 2013-2023
selected_files = [f for f in all_files if any(str(year) in f for year in range(2013, 2024))]
print(selected_files)  # Check the filtered list
```

```python
import xarray as xr
import os

data_dir = "ChaosBench/data/era5"

# Select any one file (preferably the latest)
sample_file = os.path.join(data_dir, "era5_full_1.5deg_20130101.zarr")

# Open dataset
ds = xr.open_zarr(sample_file, consolidated=True)

# Print all available variables
print("Available Variables in Dataset:\n", list(ds.variables))
```

```python
import xarray as xr

data_dir = "ChaosBench/data/era5"

# Open multiple .zarr files and merge them
datasets = []
for file in selected_files:
    ds = xr.open_zarr(os.path.join(data_dir, file), consolidated=True)
    datasets.append(ds)

# Merge into one dataset
merged_ds = xr.concat(datasets, dim="time")

# Check dimensions
print(merged_ds)
```

We combined the internal folders in dataset and got the required variables, timelines and locations by slicing

```python
# Define pressure levels we need
required_levels = [850, 500]  # These correspond to 850hPa and 500hPa

# Filter dataset to keep only the required pressure levels
filtered_ds = merged_ds.sel(level=required_levels)
```

```python
# Define the variable mapping
#selected_vars = ['z', 't', 'q', 'u', 'v']  # Available variables in dataset
selected_vars = ['z', 't']  # Only two variable for new dataset

# Select only required variables
filtered_ds = filtered_ds[selected_vars]
```

```python
# Define our Midwest U.S. lat/lon range
lat_range = slice(50, 30)   # Midwest U.S. latitudes (reversed because xarray)
lon_range = slice(255, 275) # Midwest U.S. longitudes

# Define our time range (2013-2023)
time_range = slice("2013-01-01", "2023-12-31")

# Apply filtering
filtered_ds = filtered_ds.sel(time=time_range, latitude=lat_range, longitude=lon_range)
```

```python
print(filtered_ds)
```

```python
import xarray as xr

# Create a dictionary to store the new variables
renamed_vars = {}

# Loop through selected variables and levels, renaming them
for var in ['z', 't']:  # Available variables in dataset
    for level in [500, 850]:  # Pressure levels we need
        new_var_name = f"{var}{level}"  # Example: z500, z850
        renamed_vars[new_var_name] = filtered_ds[var].sel(level=level).drop_vars("level")

# Convert dictionary to a new dataset
final_era5_ds = xr.Dataset(renamed_vars)

# Select only the required variables for new dataset
final_era5_ds = final_era5_ds[['z500', 't850']]

# Display final dataset structure
print(final_era5_ds)

```

```python
# Get the unique latitude and longitude values
num_latitudes = len(final_era5_ds.latitude)
num_longitudes = len(final_era5_ds.longitude)

# Compute total unique grid points
num_regions = num_latitudes * num_longitudes

print(f"Total unique grid points (regions): {num_regions}")
print(f"Latitude points: {num_latitudes}, Longitude points: {num_longitudes}")
```

```python
# Save as NetCDF for further processing
final_era5_ds.to_netcdf("/content/final_era5_chaosbench_data.nc")
```

```python
import xarray as xr
import pandas as pd

# Load the NetCDF file
nc_file = "/content/final_era5_chaosbench_data.nc"
ds = xr.open_dataset(nc_file)

# Display dataset structure
print(ds)
```

```python
# Convert to pandas DataFrame
df = ds.to_dataframe().reset_index()

# Display the first few rows
print(df.head())
df.count()
```

```python
csv_file_path = "/content/final_era5_chaosbench_data.csv"
df.to_csv(csv_file_path, index=False)

print(f"CSV file saved at: {csv_file_path}")
```

```python
import os

data_dir2 = "ChaosBench/data/lra5"
all_files = sorted(os.listdir(data_dir2))  # List all available files

# Filter only the files for 2013-2023
selected_files2 = [f for f in all_files if any(str(year) in f for year in range(2013, 2024))]
print(selected_files2)  # Check the filtered list
```

```python
import xarray as xr
import os

data_dir2 = "ChaosBench/data/lra5"

# Select any one file (preferably the latest)
sample_file2 = os.path.join(data_dir2, "lra5_full_1.5deg_19790101.zarr")

# Open dataset
ds2 = xr.open_zarr(sample_file2, consolidated=True)

# Print all available variables
print("Available Variables in Dataset:\n", list(ds2.variables))
```

```python
import xarray as xr

data_dir2 = "ChaosBench/data/lra5"

# Open multiple .zarr files and merge them
datasets = []
for file in selected_files2:
    ds2 = xr.open_zarr(os.path.join(data_dir2, file), consolidated=True)
    datasets.append(ds2)

# Merge into one dataset
merged_ds2 = xr.concat(datasets, dim="time")

# Check dimensions
print(merged_ds2)
```

```python
# Define our Midwest U.S. lat/lon range
lat_range = slice(50, 30)   # Midwest U.S. latitudes (reversed in xarray)
lon_range = slice(255, 275) # Midwest U.S. longitudes

# Define our time range (2013-2023)
time_range = slice("2013-01-01", "2023-12-31")

# Define missing variables
#selected_vars_lra5 = ['t2m', 'd2m', 'u10', 'v10', 'tp', 'pev', 'swvl1', 'ssrd', 'lai_hv', 'lai_lv'] # Old dataset avriables

selected_vars_lra5 = [ 'e', 'evavt', 'lai_lv', 'pev', 'swvl2', 'swvl3', 't2m', 'tp'] #

# Apply filtering
filtered_ds_lra5 = merged_ds2.sel(time=time_range, latitude=lat_range, longitude=lon_range)[selected_vars_lra5]
```
